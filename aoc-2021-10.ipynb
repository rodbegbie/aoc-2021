```python
# SAMPLE 

INPUT = """[({(<(())[]>[[{[]{<()<>>
[(()[<>])]({[<{<<[]>>(
{([(<{}[<>[]}>{[]{[(<()>
(((({<>}<{<{<>}{[]{[]{}
[[<[([]))<([[{}[[()]]]
[{[{({}]{}}([{[{{{}}([]
{<[[]]>}<{[{[{[]{()[[[]
[<(<(<(<{}))><([]([]()
<{([([[(<>()){}]>(<<{{
<{([{{}}[<[[[<>{}]]]>[]]"""
```

```python
# Part 1

SYNTAX_SCORES = {
    ")": 3,
    "]": 57,
    "}": 1197,
    ">": 25137,
}

total_score = 0

for row in INPUT.splitlines():
    print ("ROW", row)
    open_chunks = ""

    for char in row:
        # print ("CHAR", char)
        if char == "(":
            open_chunks += ")"
        elif char == "<":
            open_chunks += ">"
        elif char == "{":
            open_chunks += "}"
        elif char == "[":
            open_chunks += "]"
        else:
            if open_chunks[-1] != char:
                print ("SYNTAX ERROR", char)
                total_score += SYNTAX_SCORES[char]
                break
            open_chunks = open_chunks[:-1]

print ("ANSWER", total_score)
```

```python
# Part 2

INCOMPLETE_SCORES = {
    ")": 1,
    "]": 2,
    "}": 3,
    ">": 4,
}

incomplete_scores = []

for row in INPUT.splitlines():
    print ("ROW", row)
    open_chunks = ""

    for char in row:
        # print ("CHAR", char)
        if char == "(":
            open_chunks += ")"
        elif char == "<":
            open_chunks += ">"
        elif char == "{":
            open_chunks += "}"
        elif char == "[":
            open_chunks += "]"
        else:
            if open_chunks[-1] != char:
                print ("SYNTAX ERROR", char)
                open_chunks = ""
                break
            open_chunks = open_chunks[:-1]

    if open_chunks:
        print ("ADDING", open_chunks)
        total_score = 0
        for char in reversed(open_chunks):
            total_score *= 5
            total_score += INCOMPLETE_SCORES[char]
        incomplete_scores.append(total_score)

print ("SCORES", incomplete_scores)
print ("ANSWER", sorted(incomplete_scores)[len(incomplete_scores)//2])
```
