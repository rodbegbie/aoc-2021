```python
# PART 1
from collections import defaultdict

lines = INPUT.splitlines()
template = lines.pop(0)
lines.pop(0)

insertions = {}
for line in lines:
    (pair, insertion) = line.split(" -> ")
    insertions[pair] = pair[0] + insertion

for i in range(10):
    print (i, template[:20], len(template))
    new_template = ""
    for i in range(len(template)):
        pair = template[i:i+2]
        if pair not in insertions:
            new_template += pair[0]
        else:
            new_template += insertions[pair]

    template = new_template

char_counts = defaultdict(int)
for char in template:
    char_counts[char] += 1

print (char_counts)
max_char_count = max(char_counts.values())
min_char_count = min(char_counts.values())

print (max_char_count, min_char_count, max_char_count - min_char_count)
```

```python
# PART 2

from collections import defaultdict

lines = INPUT.splitlines()
template = lines.pop(0)
lines.pop(0)

TARGET_DEPTH = 40

insertions = {}
for line in lines:
    (pair, insertion) = line.split(" -> ")
    insertions[pair] = pair[0] + insertion

depth_map = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))

def get_insertion(pair):
    return insertions.get(pair, pair)

def get_pairs(template):
    pairs = []
    for i in range(len(template)):
        pairs.append(template[i:i+2])
    return pairs

def get_letter_count(pair, depth):
    # print ("pair", pair, "depth", depth)
    if depth_map[pair][depth]:
        # print ("FOUND!", depth_map[pair][depth])
        return depth_map[pair][depth]
    elif len(pair) == 1:
        depth_map[pair][0][pair[0]] = 1
        return depth_map[pair][0]
    elif depth == 0:
        depth_map[pair][0][pair[0]] = 1
        depth_map[pair][0][pair[1]] += 1
        # print ("NEW!", depth_map[pair][depth])
        return depth_map[pair][0]
    else:
        inserted = get_insertion(pair) + pair[1]
        for p in get_pairs(inserted)[:-1]:
            one_higher = get_letter_count(p, depth-1)
            for (letter, value) in one_higher.items():
                depth_map[pair][depth][letter] += value
        depth_map[pair][depth][inserted[1]] -= 1
        
        # print ("RET", pair, depth,  depth_map[pair][depth])
        return depth_map[pair][depth]

final_map = defaultdict(int)
for p in get_pairs(template):
    one_higher = get_letter_count(p, TARGET_DEPTH)
    for (letter, value) in one_higher.items():
        final_map[letter] += value
    if len(p) > 1:
        final_map[p[1]] -= 1

max_char_count = max(final_map.values())
min_char_count = min(final_map.values())

print (max_char_count, min_char_count, max_char_count - min_char_count)            
```
