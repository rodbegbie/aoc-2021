```python
# Part 1

hor = dep = 0

for row in INPUT.splitlines():
    (instruction, number) = row.split()
    number = int(number)
    if instruction == "forward":
        hor += number
    elif instruction == "down":
        dep += number
    else:
        dep -= number

print ("POSITION", hor*dep)
```
