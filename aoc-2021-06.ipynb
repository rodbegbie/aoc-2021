```python
# Part 1

lanternfish = [int(x) for x in INPUT.split(",")]
day = 1

while day <= 80:
    oldfish = []
    newfish = []
    for fish in lanternfish:
        if fish == 0:
            oldfish.append(6)
            newfish.append(8)
        else:
            oldfish.append(fish - 1)
    lanternfish = oldfish + newfish

    print ("DAY", day, "NUMBER", len(lanternfish))
    day += 1
```

```python
# Part 2

from collections import defaultdict

lanternfish = [int(x) for x in INPUT.split(",")]
day = 1

fish_counts = defaultdict(int)
for fish in lanternfish:
    fish_counts[fish] += 1

while day <= 256:
    new_fish_counts = defaultdict(int)
    new_fish_counts[6] = fish_counts[0]
    new_fish_counts[8] = fish_counts[0]
    for i in range(1,9):
        new_fish_counts[i-1] += fish_counts[i]

    fish_counts = new_fish_counts
    print ("DAY", day, "NUMBER", sum(fish_counts.values()))
    day += 1
```
