```python
# Part 1
GRID = [[int(x) for x in line] for line in INPUT.splitlines()]
step = 0
flashes = 0

def print_grid():
    print ("\nSTEP", step)
    print ("\n".join(["".join([str(x) if x > 0 else "•" for x in line]) for line in GRID]))

# print_grid()

while step < 100:
    step += 1
    flashed = set()

    for y in range(10):
        for x in range(10):
            GRID[y][x] += 1

    while True:
        to_flash = set()
        for y in range(10):
            for x in range(10):
                if (x,y) not in flashed:
                    if GRID[y][x] > 9:
                        to_flash.add((x,y))

        if not to_flash:
            break

        for (flash_x, flash_y) in to_flash:
            GRID[flash_y][flash_x] = 0
            flashed.add((flash_x, flash_y))
            flashes += 1
            for inc_x in (-1, 0, 1):
                for inc_y in (-1, 0, 1):
                    consider_x = flash_x + inc_x
                    consider_y = flash_y + inc_y
                    if (consider_x, consider_y) in flashed or \
                        (consider_x, consider_y) == (flash_x, flash_y):
                        continue
                    if 0 <= consider_x <= 9 and 0 <= consider_y <= 9:
                        GRID[consider_y][consider_x] += 1
    print_grid()

print ("ANSWER", flashes)


```
