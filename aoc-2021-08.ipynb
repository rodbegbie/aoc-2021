```python
# Part 2

display_numbers = [[[set(number) for number in chunk.split(" ")] for chunk in line.split(" | ")] for line in INPUT.splitlines()]
total = 0
for (input, output) in display_numbers:
    for number in input:
        if len(number) == 2:
            one = number
        elif len(number) == 3:
            seven = number
        elif len(number) == 4:
            four = number
        elif len(number) == 7:
            eight = number
        
    A = seven - one
    # print ("A", A)

    # Find 9 to get E & G
    seven_and_four = seven | four
    for number in input:
        if len(number) == 6:
            diff = number - seven_and_four
            if len(diff) == 1:
                nine = number
                G = nine - seven_and_four
                E = eight - nine
                # print ("E", E, "G", G)
    
    # Find D
    five_segments = []
    for number in input:
        if len(number) == 5:
            five_segments.append(number)
            if E & number:
                two = number
    
    D = (five_segments[0] & five_segments[1] & five_segments[2]) - A - G
    C = two - (A | D | G | E)
    # print ("D", D, "C", C)

    zero = eight - D
    F = one - C
    B = four - (C | D | F) 

    three = (A | C | D | F | G)
    five = (A | B | D | F | G)
    six = (A | B | D | E | F | G)
    
    # print (A, B, C, D, E, F, G)

    NUMBERS = (zero, one, two, three, four, five, six, seven, eight, nine)

    def get_int(number):
        answer = 0
        for i in NUMBERS:
            if not i ^ number:
                return answer
            answer += 1
    
    total += (
        (1000 * get_int(output[0])) +
        (100 * get_int(output[1])) +
        (10 * get_int(output[2])) +
        get_int(output[3])
    )

print ("TOTAL", total)


```
